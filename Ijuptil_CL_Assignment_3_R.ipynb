```r
Assignment
```

```r
for (i in (1:100)){
    
    sort1 <- c(Assignment$Ex1_30[i], Assignment$Ex2_30[i], Assignment$Ex3_30[i], Assignment$Ex4_30[i], Assignment$Ex5_30[i] )
    sort2 <- sort(sort1, decreasing = TRUE)
    Best_Ex <- round(sum(sort2[1:3])/3, digits=1)
    Assignment$Av_B_EX[i] <-  Best_Ex
}       # This command is to choose the best three exercises and find the average of the best three
```

```r
Assignment
```

```r
Assignment['Exercises'] =round((Assignment$Av_B_EX/30)*10, digits =1)
```

```r
Assignment
```

```r
Assignment['Total_Text'] = ((Assignment$Test1_20 + Assignment$Test2_20)/40)*20

Assignment['Project'] = ((Assignment$Project_60)/60)*30

Assignment['CA'] = round((Assignment$Exercises + Assignment$Total_Text), digits =1)

Assignment['Exam_40'] = ((Assignment$Exam_100)/100)*40

Assignment['Total'] = round(Assignment$Project + Assignment$CA + Assignment$Exam_40, digits =0)
```

```r
Assignment
```

```r
StudentGrades <- function(x){
    if (is.vector(x)){
        result = rep(NaN,length(x))
        result[(x >= 80) & (x<=100)] = 'A'
        result[(x >= 75) & (x < 80)] = 'B+'
        result[(x >= 70) & (x < 75)] = 'B'
        result[(x >= 65) & (x < 70)] = 'C+'
        result[(x >= 60) & (x < 65)] = 'C'
        result[(x >= 55) & (x < 60)] = 'D+'
        result[(x >= 50) & (x < 55)] = 'D'
        result[(x >= 45) & (x < 50)] = 'E'
        result[(x >= 0) & (x < 45)] = 'F'
    }
        
    return(result)
}
```

```r
Assignment$Grades = StudentGrades(Assignment$Total)

table(Assignment$Grades, useNA = 'ifany')
Assignment
```

```r
cor(Assignment$CA, Assignment$Exam_40) # This line of code is to find the correlation between CA and Exam
```

```r
model <- lm(Assignment$Exam_40 ~ Assignment$CA ) # This command is for linear regression model
```

```r
summary(model) # This command displays the summary statistics of the model
```

```r
coef(model)
```

```r
plot(Assignment$CA, Assignment$Exam_40, main ="Graph of Continuous Assessment against Exam", col= "red", lwd = 3) + abline(model, col ='blue', lwd = 2)
```
