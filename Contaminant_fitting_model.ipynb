```python
import numpy as np
from scipy.integrate import odeint
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit


# parameters
cin = 2840
thalf= 98.7

# Define the analytical function
def analytical_function(t, Kad):
    return cin/(1 + np.exp((Kad/10000)*cin*(thalf-t)))

# Observed data
observed_data_file = '/content/data_clean2835.csv'
observed_data = pd.read_csv(observed_data_file)

# Extract t and C
tarray = observed_data['t (min)']
Carray = observed_data['C (mg/m3)']

# Scatter pot t and C
plt.scatter(tarray, Carray, label='observed_C(t)')

# Perform parameter fitting
popt, pcov = curve_fit(analytical_function, tarray, Carray, p0=[1.3])

# Plot the fitted curve
t_fit = np.linspace(min(tarray), max(tarray), 50)
C_fit = analytical_function(t_fit, *popt)
plt.plot(t_fit, C_fit, 'r-', label='Fitted_C(t)')

plt.xlabel('t [min]')
plt.ylabel('C [mg/m3]')
plt.legend()
plt.title('Contaminant Problem')
plt.show()

# Print the optimized parameters
print("Optimized Parameters:")
print("Kad =", popt[0])

#
# Combine x_fit and C_fit into a single array
data = np.column_stack((t_fit, C_fit))

# Save the data to a CSV file
np.savetxt('ccap_data.csv', data, delimiter=',', header='t (min), Ct', comments='')

print("ccap_data.csv")

```
