```python
import re
def N_Gram(n, s):
    tokens = re.split("\\s+", s)   # split sentence into token
    ngrams = []
    for i in range(len(tokens)-n+1):    # This is to collect the number of n grams
        L = [tokens[j] for j in range(i, i+n)]
        ngrams.append("".join(L))
    return set(ngrams)
```

```python
def RungeKutta(f0, a, b, y0, h):
    n = int((b-a)/h)
    t = np.array([a*i+h for i in range(n+1)])
    y2 = [y0]
   
   
    for i in range(n):
        k1 = h*f0(t[i-1],y2[i-1])
        k2 = h*f0(t[i-1] + h/2,y2[i-1]+k1/2)
        k3 = h*f0(t[i-1] + h/2,y2[i-1] + k2/2)
        k4 = h*f0(t[i-1] + h,y2[i-1] + k3)
        y2_i = y2[i-1] + k1/6 + k2/3 + k3/3 + k4/6
        y2.append(y2_i)
    return t,y2

def f0(t,y2):
    r = 0.4
    k = 20
    return r*y2*(1 - y2/k)

t,y2 = RungeKutta(f0, 0.0, 50, 2.44, 0.01)
print(y2,end=' ')
```
